```python
# lets split hte colunms into X and Y features
X=g_t.drop(columns=['TEY'],axis=1)
X
```

```python
Y=g_t.iloc[:,[7]]
Y
```

```python
# lets check for the relu function
from keras.models import Sequential
from keras.layers import Dense,Activation
#from keras.optimizers import Adam
model=Sequential()
model.add(Dense(10,input_dim=10,kernel_initializer='uniform',activation='relu'))
model.add(Dense(4,kernel_initializer='uniform',activation='relu'))
model.add(Dense(4,kernel_initializer='uniform',activation='relu'))
#adam=Adam()
model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
model.summary()
```

```python
#lets compile the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#lets fit the model
history=model.fit(X,Y,validation_split=0.20,epochs=100,batch_size=10)
```

```python
#lets evaluate the model
scores=model.evaluate(X,Y)
print("%s: %0.2f%%" %(model.metrics_names[1],scores[1]*100))
```

```python
#since Y value is continous we vl use regression model
# deofne the classifiaction model
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV,KFold,cross_val_score

def reg_model():
    model=Sequential()
    model.add(Dense(10,input_dim=10,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
    model.add(Dense(8,kernel_initializer='uniform',activation='relu'))
   # adam=Adam()
    model.add(Dense(1,kernel_initializer='uniform',activation='sigmoid'))
    #model.add(Dense(1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model


regr=KerasRegressor(build_fn=reg_model,verbose=0)
batch_size=[10,20,40,100]
epochs= [10,20,50,100]
grid_param= dict (batch_size=batch_size,epochs=epochs)
grid=GridSearchCV(estimator=regr,cv=KFold(),param_grid=grid_param,verbose=10)
grid_results=grid.fit(X,Y)
```

```python
print('best: {},score:{}'.format(grid_results.best_score_,grid.best_params_))
```
